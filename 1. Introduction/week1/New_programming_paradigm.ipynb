```python
from tensorflow import keras
import numpy as np

# 신경망(모델) 구성
# 1 Dense, 1 Unit, 1 Input_shape
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])

# optimizer(최적화 함수) : sgd(Stochastic Gradient Descent), 확률적 경사 하강법
# loss(손실 함수) : mean_squared_error, 평균 제곱 오차
model.compile(optimizer='sgd', loss='mean_squared_error')

# X, Y 데이터
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

# 훈련단계, epochs : 전체 데이터 셋에 대해 학습 횟수 단위.
model.fit(xs, ys, epochs=100)

print(model.predict([10.0]))
```


**epochs = 100에 해당하는 예측값(17.54539)이 19가 아닌 이유**

1. 적은 데이터로 학습 했기 때문에. 모든 x값에 대해서 y가 예측값과 같을 것이라는 보장이 없다.
2. 신경망을 사용할 때 학률로 다룬다.따라서, 예측 결과를 처리하는 방법을 조정해야한다.
